```python
df_yahoo=pd.read_csv("yahoo_screening_nasdaq_20190830.csv")
```

```python
undervalued=df_yahoo.loc[df_yahoo['overvalued']=='Near Fair Value']
undervalued_buy=undervalued.loc[undervalued['buy']=='BUY']
undervalued_buy
```

```python
tickers=undervalued_buy.name.to_list()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException


l=[]
browser = webdriver.Chrome()
browser_buy = webdriver.Chrome()
browser_buy.set_page_load_timeout(120)
browser_zack = webdriver.Chrome()

y=[]
for ticker in tickers:
    site= 'https://finance.yahoo.com/quote/' +ticker+ '/key-statistics/'
    browser.get(site)
    site_buy= 'https://finance.yahoo.com/quote/' +ticker
    browser_buy.get(site_buy)
    site_zack= 'https://www.zacks.com/stock/quote/' +ticker+ '/financial-overview'
    browser_zack.get(site_zack)
    # get all the financial values
    e={}
    e['name']=ticker
    try:
        e['peg'] = browser.find_element_by_xpath('//*[@id="Col1-0-KeyStatistics-Proxy"]/section/div[2]/div[1]/div/div/div/table/tbody/tr[5]/td[2]').text
    except:
        pass
    try:
        e['price_to_sale'] = browser.find_element_by_xpath('//*[@id="Col1-0-KeyStatistics-Proxy"]/section/div[2]/div[1]/div/div/div/table/tbody/tr[6]/td[2]').text
    except:
        pass
    try:
        e['price_to_book'] = browser.find_element_by_xpath('//*[@id="Col1-0-KeyStatistics-Proxy"]/section/div[2]/div[1]/div/div/div/table/tbody/tr[7]/td[2]').text
    except:
        pass
    try:
        e['50_dma'] = browser.find_element_by_xpath('//*[@id="Col1-0-KeyStatistics-Proxy"]/section/div[2]/div[2]/div/div[1]/div/table/tbody/tr[6]/td[2]').text
    except:
        pass
    try:
        e['200_dma'] = browser.find_element_by_xpath('//*[@id="Col1-0-KeyStatistics-Proxy"]/section/div[2]/div[2]/div/div[1]/div/table/tbody/tr[7]/td[2]').text
    except:
        pass
    try:
        e['short_ratio'] = browser.find_element_by_xpath('//*[@id="Col1-0-KeyStatistics-Proxy"]/section/div[2]/div[2]/div/div[2]/div/table/tbody/tr[8]/td[2]').text
    except:
        pass
    try:
        e['overvalued'] = browser_buy.find_element_by_xpath('//*[@id="rsrch-insghts"]/div/div[2]/div[2]/div[1]').text
    except:
        pass
    try:
        e['buy'] = browser_buy.find_element_by_xpath('//*[@id="rsrch-insghts"]/div/div[1]/div[2]/span').text
    except:
        pass
    try:
        e['zack_rank'] = browser_zack.find_element_by_xpath('//p[@class="rank_view"]').text.split('-')[0]
    except:
        pass
    try:
        e['vgm_score'] = browser_zack.find_element_by_xpath('//*[@id="quote_ribbon_v2"]/div[2]/div[3]/p/span[4]').text
    except:
        pass

    y.append(e)
#save to dataframe
import pandas as pd
df_yahoo=pd.DataFrame(y)
df_yahoo
# df_yahoo.to_csv("yahoo_screening_nasdaq_20190830.csv")
```
